```python
plt.scatter(data=pokemon,x='weight_kg',y='height_m',s=500,marker="*",c='orange')
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.scatter(data=pokemon,x='sp_attack',y='sp_defense',s=250,marker="*",c='orange')
#Change axis length
plt.xlim(0,250)
plt.ylim(0,250)
#plt title
plt.title("Relationship b/w Attack and Defense",size=20)
plt.xlabel("attack",size=15)
plt.ylabel("defense",size=15)
plt.grid(True,linestyle='--')
#Saving a file
plt.savefig('Attack_Defense.jpg')
plt.show()
```

```python
import seaborn as sns
```

```python
#Scatterplot with regression line
sns.lmplot(data=pokemon,x='sp_attack',y='sp_defense')
```

```python
sns.lmplot(data=pokemon,x='attack',y='defense',hue='is_legendary',palette='Set1') #,palette='seismic'
```

```python
#filtering a data
sns.lmplot(data=pokemon.loc[pokemon['is_legendary']==1],x='attack',y='defense',hue='is_legendary',fit_reg=False,palette='Set1') #,palette='seismic'
```

```python
sns.pairplot(data=pokemon,x_vars=['attack','defense','speed'],y_vars='base_total',hue='is_legendary')
```

```python
#line plot
X=np.arange(0,10)
Y=np.arange(11,21)
#plt.plot(pokemon['attack'],pokemon['speed'],c='green')
plt.plot(X,Y,c='green')
for i_x,i_y in zip(X,Y):
    plt.text(i_x,i_y,(i_x,i_y))

plt.show()
```

```python
plt.figure(figsize=(20,10))

plt.subplot(131)
plt.hist(data=pokemon,x='attack',color='darkblue',bins=20)
#plt title
plt.title("Histogram Attack",size=20)
plt.grid(True,linestyle='--')

plt.subplot(132)
plt.hist(data=pokemon,x='defense',color='blue',bins=10)
#plt title
plt.title("Histogram Defense",size=20)
plt.grid(True,linestyle='--')

plt.subplot(133)
plt.hist(data=pokemon,x='speed',color='orange',bins=10)
#plt title
plt.title("Histogram Speed",size=20)
plt.grid(True,linestyle='--')

#Saving a file
plt.savefig('Histogrm_Attack_Defense.jpg')
plt.show()
```
