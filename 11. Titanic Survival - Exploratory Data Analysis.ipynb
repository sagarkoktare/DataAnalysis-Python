```python
train['Age'] = train['Age'].fillna(round(train['Age'].mean()))
```

```python
train['Cabin'] = train['Cabin'].fillna(train['Cabin'].mode()[0])
train['Embarked'] = train['Embarked'].fillna(train['Embarked'].mode()[0])
```

```python
train.describe(include="all")
```

```python
train.dropna(how='any',subset=['Age','Cabin'])
train.shape
```

```python
#50% threshold
half_count = len(train)/2
#75% threshold
half_count = len(train) * 0.75

print(half_count)
train.dropna(thresh = half_count, axis = 1)
train.shape
```

```python
#Discretization and Binning
P_age = train['Age']
Bins = [0,15,21,60,P_age.max()]
BinLabels = ['Childrens','Young','Adults','Seniors']
categories = pd.cut(P_age,Bins,labels=BinLabels)
```

```python
categories
```

```python
train = pd.concat([train,categories],axis=1)
```

```python
train.head()
```

```python
train.columns=['Survived', 'Pclass', 'Name', 'Sex', 'Age', 'SibSp', 'Parch', 'Ticket',
       'Fare', 'Cabin', 'Embarked', 'Age bands']
```
